# Linear 3 dimensional Environment, sd level 0, 40 prompts 10

```python
save_folder = '../results/plots/supplement/fogambitt_linear_env_40_prompts_10k_horizon'
if not os.path.exists(save_folder):
    os.makedirs(save_folder, exist_ok=True)
```

```python
results = load_simulation_results('../data/results/fogammbit_linear_env_40_2025-10-14', env="linear", sds = ["sd_0"]) 
```

```python
agg_cols = ['agent_type', "t"]
agg_res = get_agg_results(df=results, group_ix=agg_cols) 
sim_cr = agg_res.cum_regret.reset_index()  
```

```python
p = ( make_regret_plot_matplotlib(df = sim_cr, 
fill = "agent_type", 
title = ""))
with open(os.path.join(save_folder, 'foGMB_linear_sd0_40_prompts_10k_horizon.pckl'), "wb") as f:
    pickle.dump(p, f)
```

```python
p = ( make_regret_plot_matplotlib(df = sim_cr, 
fill = "agent_type", 
title = "",
include_legend=False))
with open(os.path.join(save_folder, 'foGMB_linear_sd0_40_prompts_10k_horizon_nolgnd.pckl'), "wb") as f:
    pickle.dump(p, f)
```
