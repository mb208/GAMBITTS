```python
from matplotlib.legend import Legend

# --- assumes you already have your plotted figure ---
legends = []
for ax in fig.axes:
    for art in ax.artists:
        if isinstance(art, Legend):
            legends.append(art)

print(f"Found {len(legends)} legends")

# --- Save each legend as a separate PDF ---
for i, L in enumerate(legends, start=1):
    # Use the public accessors
    handles = list(L.legend_handles)
    labels  = [t.get_text() for t in L.texts]
    title   = L.get_title().get_text() or f"legend_{i}"

    # Create new standalone figure for this legend
    fig_legend = plt.figure(figsize=(4, 2), constrained_layout=True)
    fig_legend.legend(
        handles,
        labels,
        loc="center",
        title=title,
        frameon=False,      # removes gray box
        fontsize=14,
        title_fontsize=16,
        ncol=1
    )

    # Save it in your desired folder with your naming convention
    filename = f"ponats_1d_linear_hedge_legend_only_{title.replace(' ', '_')}.pdf"
    out_path = os.path.join(png_folder, filename)
    fig_legend.savefig(out_path, bbox_inches='tight', transparent=True)
    plt.close(fig_legend)

    print(f"Saved legend to: {out_path}")
```
